# Using Newtons Method:

```python
total=6.4
# Sigmoid function
def sigmoid(z):
    return 1 / (1 + np.exp(-z))
```

```python
class LogisticRegression:
    
    def __init__(self, learning_rate=0.1, num_iterations=1000, alpha=0.1):
        
        self.learning_rate = learning_rate
        self.num_iterations = num_iterations
        self.alpha = alpha  # L2 regularization parameter

    def fit(self, X, y):
        m, n = X.shape
        self.theta = np.zeros(n)
        
        for i in range(self.num_iterations):
            z = np.dot(X, self.theta)
            h = sigmoid(z)
            gradient = (np.dot(X.T, (h - y)) + 2 * self.alpha * self.theta) / m  
            self.theta -= self.learning_rate * gradient

    def predict(self, X):
        z = np.dot(X, self.theta)
        return (sigmoid(z) >= 0.5).astype(int)

    def accuracy(self, y_true, y_pred):
        return np.mean(y_true == y_pred)
```

```python
#train the model
model = LogisticRegression(learning_rate=0.1, num_iterations=1000, alpha=0.1)
model.fit(X_train, y_train)
```

```python
# Make predictions on test data
y_pred = model.predict(X_test)
```
